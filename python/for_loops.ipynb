```python
#for loop
colors = ["red","green","blue", "yellow"]

for color in colors:
    print(color)
```

```python
#loop with multiple loop variables
points = [(1,2), (3,4), (5,6)]

for x, y in points:
    print(f"x: {x}, y: {y}")
```

```python
# The target iterable is an empty list. The loop checks whether the iterable has items. If that’s the case, then the loop runs once for each item. If the iterable has no items, then the loop body doesn’t run, and the program’s execution flow jumps onto the statement after the loop.

for item in []:
    print(item)
```

```python
#Tuple iteration

person = ("Alice", 30, "Engineer")
for attribute in person:
    print(attribute)
```

```python
#String character iteration

text = "abcde"

for char in text:
    print(char)
    
```

```python
#Range
for i in range(5):
    print(i)
```

```python
students = {
    "Alice": 89.5,
    "Bob": 76.0,
    "Charlie": 92.3,
    "Diana": 84.7,
    "Ethan": 88.9
}

for student in students.keys():
    print(student)
```

```python
students = {
    "Alice": 89.5,
    "Bob": 76.0,
    "Charlie": 92.3,
    "Diana": 84.7,
    "Ethan": 88.9
}

for student in students.values():
    print(student)
```

```python
# .items() method returns a view object that displays a list of a dictionary's key-value tuple pairs. This allows you to iterate over both keys and values simultaneously in a for loop.
students = {
    "Alice": 89.5,
    "Bob": 76.0,
    "Charlie": 92.3,
    "Diana": 84.7,
    "Ethan": 88.9
}

for student, marks in students.items():
    print(student, "->", marks)
```

```python
#When it comes to iterating over sets, you only have to keep in mind that sets are unordered data types. This means that the order in which you get the items when iterating over a set is not guaranteed to be the same as the order in which you added them to the set. However, you can still use a for loop to iterate over a set and access its items one by one.

tools = {"hammer", "screwdriver", "wrench"}

for tool in tools:
    print(tool)
```

```python
#The break Statement

numbers = [1, 2, 3, 4, 5]
target = 5

for number in numbers:
    if number == target:
        print("Target found!")
        break
    print(number)
```

```python
#The continue Statement

numbers = [1, 2, 3, 4, 5, 6]

for number in numbers:
    if number % 2 != 0:
        continue
    print(number)
```

```python
#The else Clause in For Loops

numbers = [1, 3, 5, 7, 9, 42]
target = 42

for number in numbers:
    print(f"Checking number: {number}")
    if number == target:
        print("Target found!")
        break
else:
    print("Target not found in the list.")
```

```python
#The enumerate() function takes an iterable as an argument and generates tuples of the form (index, item)
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")
```
