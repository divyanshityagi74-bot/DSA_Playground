```python
# count all subsets of a set of size N
# I/P : { "n" : 3}    O/P : 8 as =>  {}, {1}. {2}, {1,2}, {3}, {1,3}, {2,3}, {1,2,3}
```

```python
def count(n):
    if n == 0:
        return 1
    else:
         return 2 * count(n-1)
    
print(count(4))
#Time Complexity - O(n)
```

```python
#Optimal Solution

def count(n):
    if (n==0):
        return 1
    prod = count(n//2)
    if (n%2 == 0):
        return prod * prod
    else: 
        return 2 * prod * prod

print(count(3))
```
