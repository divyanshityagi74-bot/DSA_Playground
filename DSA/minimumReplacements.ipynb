```python
"""
Ques - For each word find the minimum replacements so that no two adjacent characters are the same.

words = ["ab", "aab", "abb", "abab", "abaaaba"]
output = [0,1,1,0,1]

Return an array where each element represents number of replacements needed that word.

"ab"      → a≠b              → 0 changes ✅
"aab"     → a=a ❌           → change one 'a' → 1 change
"abb"     → b=b ❌           → change one 'b' → 1 change
"abab"    → no adjacent same → 0 changes ✅
"abaaaba" → a=a in middle ❌ → 1 change
"""
```

```python
"""Brute Force Approach:"""
def minimalOperations(words):
    result = []

    for word in words:
        count = 0
        for i in range(len(word)-1):
            if word[i] == word[i+1]:
                count += 1
        result.append(count)
    return result
words = ["ab", "aab", "abb", "abab", "abaaaba"]
print(minimalOperations(words))
```

```python
"""Count Clash Groups"""
def minimalOperations(words):
    result = []

    for word in words:
        count = 0
        i = 0
        while i < len(word) - 1:
            if word[i] == word[i + 1]:    # clash found
                count += 1
                i += 2                    # skip next — one fix handles it
            else:
                i += 1                    # no clash, move forward

        result.append(count)

    return result


# Test
words = ["ab", "aab", "abb", "abab", "abaaaba", "aaabbb", "aabbcc", "abcabc", "aaaa", "ababab", "abaabba", "abaabaa", "abaabab"]
print(minimalOperations(words))
```
